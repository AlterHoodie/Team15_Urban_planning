```python
import pickle
```

```python
with open('./init_plan_dhm.pickle','rb') as f:
    data = pickle.load(f)
```

```python
gdf = data['gdf']
```

```python
gdf.at[263, 'type'] = 14
```

```python
with open('./init_plan_dhm_test.pickle','wb') as f:
    pickle.dump({'gdf':gdf},f)
```

```python
import matplotlib.pyplot as plt

# Filter for type 4 geometries
residential = gdf[gdf['geometry'].geom_type=="Polygon"]

# Plot geometries with black edges
fig, ax = plt.subplots(figsize=(10, 10))
residential.plot(ax=ax, edgecolor="black", color="lightgreen")

# Plot indexes as annotations
for idx, row in residential.iterrows():
    # Get the centroid of each geometry to place the text label
    centroid = row.geometry.centroid
    ax.annotate(text=idx, xy=(centroid.x, centroid.y), ha='center', fontsize=8, color='blue')

# Optional: Set a title or any additional figure settings
plt.title("Residential Areas with Index Labels")
plt.show()
```
